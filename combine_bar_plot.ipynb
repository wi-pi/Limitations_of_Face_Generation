```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'serif'
```

```python
demo_verification = pd.read_csv('demographics_verification.csv',index_col=0)
surveys_verification = pd.read_csv('group_1_averages.csv',index_col=0)
```

```python
demo_verification = demo_verification.reset_index()
surveys_verification['Type'] = 'Survey'
surveys_verification = surveys_verification.rename(columns={'Pseudo-Verification':'Verification Accuracy'})
demo_verification = demo_verification.replace({'NOTMale':"Female"})
```

```python
all_results = pd.concat([surveys_verification,demo_verification])
all_results = all_results.replace({'SD2.1 Celebrities':"SDv2.1 Celebrities", 'Survey': 'Survey Responses', 'Any':'Any Demographic Pairs', 'Same':'Same Demographic Pairs'})
all_results = all_results.replace('Asian','East\nAsian')
```

```python
SAME = all_results[all_results['Type']=='Same Demographic Pairs'] 
SAME = SAME[SAME['Dataset']!='LFW']
```

```python
ax = sns.catplot(data=all_results,x='Verification Accuracy',y="Demographic",hue='Dataset',kind='bar',col='Type',order=['East\nAsian','Black','Indian','White','Female','Male'],
                 col_order=['Survey Responses', 'Same Demographic Pairs','Any Demographic Pairs'],margin_titles=True)
ax.set_titles(fontsize=50,col_template="{col_name}")
ax.set(xlim=[0.6,1],xticks=[0.6,0.8,1])
plt.subplots_adjust(wspace=0.2)
minorLocator = AutoMinorLocator()

for axs in ax.axes.flat:
    axs.tick_params(axis="x", labelsize=15)  # Adjust labelsize as needed
    axs.tick_params(axis="y", labelsize=15)  # Adjust labelsize as needed
    axs.title.set_fontsize(20)
    axs.xaxis.grid(True,which='both',color='black',linestyle='dotted',alpha=0.2) 
    axs.xaxis.set_minor_locator(minorLocator)
    
sns.move_legend(ax, "upper center", bbox_to_anchor=(0.5, -0.05),ncol=4,title=None,fontsize=15)
# ax.set_xticklabels(fontsize=15) 
ax.set_xlabels(fontsize=15)
ax.set_ylabels(fontsize=15)
ax.savefig('verification_accuracy_v1.pdf',dpi=500)
```
