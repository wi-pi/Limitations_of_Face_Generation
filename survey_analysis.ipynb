#### Some analysis on quality and edit correctness per demog

```python
import numpy as np
from scipy import stats

def get_statitical_test(df, split_column, analyzed_quantity):
    groups = []
    for group in df[split_column].unique():
        # print(group)
        groups.append(df[df[split_column] == group][analyzed_quantity])
    # t_statistic, p_value = stats.ttest_ind(*groups)
    f_statistic, p_value_anova = stats.f_oneway(*groups)
    alpha = 0.05
    if p_value_anova < alpha:
        print(f"One-way ANOVA test: At least one group mean is statistically different., f_stat = {f_statistic}, p_value = {p_value_anova}")
    else:
        print(f"One-way ANOVA test: There is no significant difference among group means., f_stat = {f_statistic}, p_value = {p_value_anova}")
    
    h_statistic, p_value_kruskal = stats.kruskal(*groups)
    if p_value_kruskal < alpha:
        print(f"Kruskal-Wallis test: At least one group median is statistically different., h =  {h_statistic}, p_value = {p_value_kruskal}")
    else:
        print(f"Kruskal-Wallis test: There is no significant difference among group medians., h =  {h_statistic}, p_value = {p_value_kruskal}")
    return p_value_anova, p_value_kruskal
```

```python
print('Average quality of source images by demographic group')
print(non_celeb_df.groupby('Demographic')['Q4.a'].mean().sort_values(ascending=False))
p_value_anova, p_value_kruskal = get_statitical_test(non_celeb_df, 'Demographic', 'Q4.a')

print('Average quality of source images by demographic group')
print(realism_celeb_df.groupby('Demographic')['Q4.a'].mean().sort_values(ascending=False))
p_value_anova, p_value_kruskal = get_statitical_test(realism_celeb_df, 'Demographic', 'Q4.a')

print('Average quality of source images by demographic group')
print(sd_celeb_df.groupby('Demographic')['Q4.a'].mean().sort_values(ascending=False))
p_value_anova, p_value_kruskal = get_statitical_test(sd_celeb_df, 'Demographic', 'Q4.a')
```

### Calculate Correlation of survey Qs and CLIP metrics

```python
import scipy.stats as stats	
def compute_corr(df):
    print('Source Image Quality correlation (+ve)\n')
    print('Q4.a image quality and src_image_clip (+ve)')
    edit_correlation, edit_p_value = stats.spearmanr(df['Q4.a'], df.src_clip_similarity)
    print(f"Spearman Correlation Coefficient: {edit_correlation}")
    print(f"P-value: {edit_p_value}" + ("      > 0.005\n" if edit_p_value > 0.005 else "\n"))

    print("------------------------------------------")
    print('Edited Image Quality correlation (+ve)\n')
    print('Q4.b image quality and trgt_image_clip (+ve)')
    edit_correlation, edit_p_value = stats.spearmanr(df['Q4.b'], df.tgt_clip_similarity)
    print(f"Spearman Correlation Coefficient: {edit_correlation}")
    print(f"P-value: {edit_p_value}" + ("      > 0.005\n" if edit_p_value > 0.005 else "\n"))

    print('Q4.b image quality and Edit_image_clip (+ve)')
    edit_correlation, edit_p_value = stats.spearmanr(df['Q4.b'], df.clip_edit_image_similarity)
    print(f"Spearman Correlation Coefficient: {edit_correlation}")
    print(f"P-value: {edit_p_value}" + ("      > 0.005\n" if edit_p_value > 0.005 else "\n"))

    print("------------------------------------------")
    print('Presrving same identity correlation (+ve)')
    print('Q1 same identity and 2-images Clip similarity (+ve)')
    id_correlation, id_p_value = stats.spearmanr(df.Q1_int, df.clip_image)
    print(f"Spearman Correlation Coefficient: {id_correlation}")
    print(f"P-value: {id_p_value}" + ("      > 0.005\n" if id_p_value > 0.005 else "\n"))

    print('Q1 same identity and 2-images DINO similarity (+ve)')
    dino_correlation, dino_p_value = stats.spearmanr(df.Q1_int, df.dino_image_similarity)
    print(f"Spearman Correlation Coefficient: {dino_correlation}")
    print(f"P-value: {dino_p_value}" + ("      > 0.005\n" if dino_p_value > 0.005 else "\n"))

    print('Q1 same identity and 2-images l1loss (-ve)')
    l1_correlation, l1_p_value = stats.spearmanr(df.Q1_int, df.l1loss)
    print(f"Spearman Correlation Coefficient: {l1_correlation}")
    print(f"P-value: {l1_p_value}" + ("      > 0.005\n" if l1_p_value > 0.005 else "\n"))

    print('Q1 same identity and 2-images l2loss (-ve)')
    l2_correlation, l2_p_value = stats.spearmanr(df.Q1_int, df.l2loss)
    print(f"Spearman Correlation Coefficient: {l2_correlation}")
    print(f"P-value: {l2_p_value}" + ("      > 0.005\n" if l2_p_value > 0.005 else "\n"))


    print('Drop in quality and Clip_image correlation (-ve)')
    correlation, p_value = stats.spearmanr(df.d_quality, df.clip_image)
    print(f"Spearman Correlation Coefficient: {correlation}")
    print(f"P-value: {p_value}" + ("      > 0.005\n" if p_value > 0.005 else "\n"))

    print('Wrong transformations and Clip_image correlation (-ve)')
    correlation, p_value = stats.spearmanr(df.Q3_int, df.clip_image)
    print(f"Spearman Correlation Coefficient: {correlation}")
    print(f"P-value: {p_value}" + ("      > 0.005\n" if p_value > 0.005 else "\n"))

    print("------------------------------------------")
    print('Applying transformation correlation (+ve)')
    print('Q2 edit correctness and clip-directional (+ve)')
    edit_correlation, edit_p_value = stats.spearmanr(df.Q2, df.clip_direction)
    print(f"Spearman Correlation Coefficient: {edit_correlation}")
    print(f"P-value: {edit_p_value}" + ("      > 0.005\n" if edit_p_value > 0.005 else "\n"))

    print('Q2 edit correctness and trgt_image_clip (+ve)')
    edit_correlation, edit_p_value = stats.spearmanr(df.Q2, df.tgt_clip_similarity)
    print(f"Spearman Correlation Coefficient: {edit_correlation}")
    print(f"P-value: {edit_p_value}" + ("      > 0.005\n" if edit_p_value > 0.005 else "\n"))

    print('Q2 edit correctness and Edit_image_clip (+ve)')
    edit_correlation, edit_p_value = stats.spearmanr(df.Q2, df.clip_edit_image_similarity)
    print(f"Spearman Correlation Coefficient: {edit_correlation}")
    print(f"P-value: {edit_p_value}" + ("      > 0.005\n" if edit_p_value > 0.005 else "\n"))


    print('Wrong transformations and clip_direction correlation (-ve or 0)')
    correlation, p_value = stats.spearmanr(df.Q3_int, df.clip_direction)
    print(f"Spearman Correlation Coefficient: {correlation}")
    print(f"P-value: {p_value}" + ("      > 0.005\n" if p_value > 0.005 else "\n"))
    return id_correlation, id_p_value, edit_correlation, edit_p_value, dino_correlation, dino_p_value, l1_correlation, l1_p_value, l2_correlation, l2_p_value

```
